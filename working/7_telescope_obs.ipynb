```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p_sc = 0.5 # probability of sunny day given cloudy one
p_cc = 0.5
p_cs = 0.1
p_ss = 0.9

# equilibrium probabilities
p_exp_s = 0.8335
p_exp_c = 0.17
```

```python
# res = 0: changing status, res != 0 keeping that weather

def p_c(): # results of cloudy days
    res = np.random.choice([0,1])
    return res

def p_s():
    res = np.random.choice(np.arange(0,10))
    return res
```

```python
# weather = {'c': p_c, 's':p_s}
```

```python
# start on a cloudy day, implement the weather forecast. Run the simulator for N \sim 10^4 days

current = 'c'
nsun = 0.
ncloud = 0.
N = 100000
sun_cum = np.zeros(N)

for i in range(1,N):
    if current == 'c':
        result = p_c()
        if result == 0:
            current = 's'
            nsun += 1.
        else: 
            ncloud += 1.

    else:
        result = p_s()
        if result == 0:
            current = 'c'
            ncloud += 1.
        else: 
            nsun += 1.

    sun_cum[i] = nsun/i

    
```

```python
print(nsun/N, ncloud/N)
```

```python
plt.hlines(p_exp_s,1,N+1,)
plt.plot(np.arange(1,N+1)[5000:], sun_cum[5000:])
```
