```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
deaths = np.arange(0,5)
groups = [109, 65, 22, 3, 1]
```

```python
plt.plot(deaths, groups/np.sum(groups), marker = 'o')
plt.grid(True)
```

```python
plt.scatter(deaths, groups/np.sum(groups), marker = 'o', label = 'originl data')
for mu in np.linspace(0.1, 1, 5):
    p = scipy.stats.poisson(mu)
    y = p.pmf(deaths)
    plt.plot(deaths, y, label = f'mu = {mu}')
plt.legend()
```

```python
plt.scatter(deaths, groups/np.sum(groups), marker = 'o', label = 'originl data')
p = scipy.stats.poisson(np.average(deaths, weights = groups/np.sum(groups) ))
y = p.pmf(deaths)
plt.plot(deaths, y, label = f'mu = {mu}', c = 'red')
plt.legend()
```
