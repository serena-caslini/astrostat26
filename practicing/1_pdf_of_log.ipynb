```python
import numpy as np
from numpy import random as rnd
import matplotlib.pyplot as plt
```

```python
N = 5000
xmin = 0.1
xmax = 10.
```

```python
x = rnd.uniform(xmin,xmax,N)
p_x = (1/(xmax-xmin))*np.ones(N)
```

```python
plt.hist(x, density = True)
plt.scatter(x, p_x, c = 'red', s=0.5, label = 'analytical p(x)')
plt.grid(True)
plt.legend()
```

```python
y = np.log10(x)
p_y = x*p_x*np.log(10)
```

```python
plt.hist(y, density=True, bins = 50)
plt.scatter(y, p_y, c = 'red', s=0.5, label = 'analitycal p(y)')
plt.legend()
plt.grid(True)
```

```python
print('means: \t\t log(x):', np.log10(np.mean(x)), '\t y:', np.mean(y))
print('medians:\t log(x):', np.log10(np.median(x)), '\t y:', np.median(y))
```
